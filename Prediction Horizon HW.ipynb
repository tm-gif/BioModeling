```python
import numpy as np
import matplotlib.pyplot as plt
```

### I hoped to show change in r by utilizing a for loop to loop through significant values of r

```python
#Logistic Growth Model
nsteps = 100 #number of steps in our lil simulation
ntraces = 10
x = np.zeros(shape = (nsteps, ntraces)) #Number of animals at every step
wiggle = .0001
r = 3.9
for r in np.arange(2.0,4.0):
    for itrace in range(ntraces):
        x[0,itrace] = .6 + np.random.rand(1)*wiggle
        for istep in range(1,nsteps): #The original number of animals
                x[istep,itrace] = r*x[istep-1, itrace]*(1-x[istep-1, itrace])

fig = plt.figure(figsize=(10,5))
#plt.plot(np.std(x,axis=1))
plt.plot(x)
plt.xlabel('Time');
plt.ylabel('Variability (sd) of trajectories');
```
