```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Throwing Stones Upwards
a = -1 #acceleration neg. bc it tries to reverse the flight of the stone
v = 5 #original speed, thrown up
x = 0 #thrown from pos 0
dt = 0.01 #numerical integration step is supposed to be small (why?)
nsteps = 1000
history = np.zeros(nsteps) #will be plotted later
for i in range(nsteps):
    v = v + a*dt
    x = x + v*dt
    history[i] = x
    
plt.plot(history);
```

```python
a = -1 #acceleration neg. bc it tries to reverse the flight of the stone
vy = 5 #UPWARDS for positive
vx = 1 #horizontal component od speed
frict = .01 # to account for friction
y = 0 #thrown from pos 0
x = 0 #thrown from pos 0
dt = 0.1 #numerical integration step is supposed to be small (why?)
nsteps = 1000
history = np.zeros(nsteps) #will be plotted later
for i in range(nsteps):
    vy = vy + a*dt - vy*frict #update vertical speed
    vx = vx - vx*frict #update horizontal speed
    
    x = x + vx*dt  #Update x position
    y = max(y + vy*dt, 0) #Update y position

    plt.plot(x,y,'k.');
    
```
