```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
*People showing one-by-one
*Once i-th person comes into the room, there are i-1 people 
*Calculate the probability that there is no overlap in birthdays
*i-1 people, i-1 days taken. So (i-1)/365 share of the year is taken
*The P of a new people not falling into the "taken" part of the year is 1-((i-1)/365)
*Each new person maintains the status quo of no overlap in BDs with 1-(i-1/365) probability
*for i=1, P(no overlap)=1
```

```python
n = 100
p = np.zeros(n) #an array of zeros
p[0] = 1
p[1] = 1
for i in range (2,n):
    p[i] = p[i-1]*(1-(i-1)/365) #multiplying 
    
#plt.plot(p); #P(no overlap)
plt.plot(1-p);
plt.xlabel('N People in Group');
plt.ylabel('P of no shared BD');
```
