## Predator/Prey Modeling

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook

n_animals = 100 # # of animals
n_steps = 10000 #steps 
diffusion=0.05 #width of normal distribution

x = np.ones(n_animals)*0.5 #x array of ones that is the size of the amount of animals there are
y = np.ones(n_animals)*0.5 #y array of the same

#PLOTTING
fig = plt.figure(figsize = (5,5))
#plt.plot(x,y,'k.'); #plot
#plt.xlim(0,1);
#plt.ylim(0,1);
#plt.show()

for t in range(n_steps): #time in steps range
    x+=np.random.randn(n_animals)*diffusion #x adds a random number in the range on animals and multiplies it by speed
    y+=np.random.randn(n_animals)*diffusion
    #Looping through all the animals and checking four characteristics 
    for i in range(n_animals):

       
        if x[i] < 0:
            x[i] = 0
        if x[i] > 1:
            x[i] = 1
            
        if y[i] < 0:
            y[i] = 0
        if y[i] > 1:
            y[i] = 1
    
    
    
    plt.cla()
    plt.plot(x,y,'k.', markersize = 5)
    plt.xlim(0,1);
    
    plt.ylim(0,1);
    fig.canvas.draw() #makes it move
    time.sleep(0.01)
    
    
```

## Better ways to limit values in a numpy vector

#### utilize numpy to check the same conditional statements

```python
%matplotlib notebook

n_animals = 100 # # of animals
n_steps = 10000 #steps 
diffusion=0.05 #width of normal distribution

x = np.ones(n_animals)*0.5 #x array of ones that is the size of the amount of animals there are
y = np.ones(n_animals)*0.5 #y array of the same

#PLOTTING
fig = plt.figure(figsize = (5,5))
#plt.plot(x,y,'k.'); #plot
#plt.xlim(0,1);
#plt.ylim(0,1);
#plt.show()

for t in range(n_steps): #time in steps range
    x+=np.random.randn(n_animals)*diffusion #x adds a random number in the range on animals and multiplies it by speed
    y+=np.random.randn(n_animals)*diffusion
    #Looping through all the animals and checking four characteristics 
    for i in range(n_animals):
        np.where((x[i] > 1), 0, 1)
        np.where((x[i] < 0), 0, 1) 
        np.where((y[i] < 0), 0, 1)
        np.where((y[i] < 1), 0, 1)
      
    
    
    
    plt.cla()
    plt.plot(x,y,'k.', markersize = 5)
    plt.xlim(0,1);
    
    plt.ylim(0,1);
    fig.canvas.draw() #makes it move
    time.sleep(0.01)
    
```
