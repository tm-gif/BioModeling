```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 20000
a = 0.6 #bunny reporduction
b = 1.2 #bunny predation
c = 0.5 
d = 1 #fox predation rate
e = 1 #hawk predation: An additional predator of bunnies

dt = 0.001

possible_ds = [0.5, 1, 1.5, 2]
possible_es = [1, 2.5, 3, 3.5]

x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2 = plt.axes()

for itrial in range(len(possible_ds)):
    d = possible_ds[itrial]
    e = possible_es[itrial]
    x[0] = 1
    y[0] = 1

    for t in range(1,nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1] - (b*x[t-1]*y[t-1]*2)) #assuming the death of bunnies increase by .5
        y[t] = y[t-1] + dt*((c*x[t-1]*y[t-1]/2) - d*y[t-1]) #assuming the growth of foxes decreases by half

    ax1 = fig1.add_subplot(1,len(possible_ds),itrial+1)
    ax1.plot(x, label='Bunnies');
    ax1.plot(y, label='Foxes');
    ax1.legend(loc='upper right');

    ax2.plot(x,y, label=d);
    plt.xlabel('Bunnies');
    plt.ylabel('Foxes');
 
ax2.legend(loc='upper right');
```
