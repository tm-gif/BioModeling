```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps =10000
npeople = 150
wealth = np.ones(shape = (nsteps,npeople))
coin = .1
trade_per_step = 10

for t in range(1, nsteps):
    wealth [t,:] = wealth[t-1,:] #keeps the changes made 
    for i in range(trade_per_step):
        p1 = np.random.randint(low = 0, high = npeople) #picks random p1 and p2
        p2 = np.random.randint(low = 0, high = npeople)

        if wealth[t,p1] > 0:
            wealth[t,p1] = wealth[t-1,p1]-coin
            wealth[t,p2] = wealth[t-1,p2]+coin

plt.figurefugsize=(10,10)
            
plt.subplot(1,2,1)
plt.hist(wealth[-1,:]) 
plt.xlabel("wealth")
plt.ylabel("count")


plt.subplot(1,2,2)
poor_one = np.argmin(wealth[-1,:])
rich_one = np.argmax(wealth[-1,:])
plt.plot(wealth[:,poor_one])
plt.plot(wealth[:,rich_one])
plt.xlabel("Time")
plt.ylabel("wealth")
plt.plot(wealth[:,0])
```

Compound Interest

```python
nsteps =10000
npeople = 150
wealth = np.ones(shape = (nsteps,npeople))
coin = .1
trade_per_step = 10
interest = 0.001

for t in range(1, nsteps):
    wealth [t,:] = wealth[t-1,:]*(1+interest) #keeps the changes made 
    for i in range(trade_per_step):
        p1 = np.random.randint(low = 0, high = npeople) #picks random p1 and p2
        p2 = np.random.randint(low = 0, high = npeople)

        if wealth[t,p1] > 0:
            wealth[t,p1] = wealth[t-1,p1]-coin
            wealth[t,p2] = wealth[t-1,p2]+coin

plt.subplot(1,2,1)
plt.hist(wealth[-1,:]) 
plt.xlabel("wealth")
plt.ylabel("count")


plt.subplot(1,2,2)
poor_one = np.argmin(wealth[-1,:])
rich_one = np.argmax(wealth[-1,:])
plt.plot(wealth[:,poor_one])
plt.plot(wealth[:,rich_one])
plt.xlabel("Time")
plt.ylabel("wealth")
plt.plot(wealth[:,0])
```
