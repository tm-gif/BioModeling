GROWTH OF BACTERA

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exponential Growth Model
nsteps = 100 #number of steps in our lil simulation
x = np.zeros(nsteps) #Number of animals at every step
x[0] = 1  #The original number of animals
r = 1.1  #makes the amount of babies grow -rate of population change -growth -die out

for i in range(1, nsteps):
    x[i] = x[i-1]*r
    
plt.plot(x);
```

```python

#Logistic Growth Model
nsteps = 100 #number of steps in our lil simulation
x = np.zeros(nsteps) #Number of animals at every step
x[0] = .6  #The original number of animals
r = 3.42 #makes the amount of babies grow -rate of population change -(growth -die out) also speed of reaction

for i in range(1, nsteps):
    x[i] = r*x[i-1]*(1-x[i-1])
    
plt.plot(x);
plt.ylim(0,1); #Force axis to be between 0 and 1
```
